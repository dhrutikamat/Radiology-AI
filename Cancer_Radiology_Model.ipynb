Check whether tumour is malignant or benign. So malignant meaning that the tumour needs to be removed and benign, meaning that the tumour is not cancerous. So we see here we have in the labelled diagnosis. 1= malignant and 0= benign. So having AI help radiologist can actually save lives.

```python
import pandas as pd

dataset = pd.read_csv('cancer.csv')
```

```python
x = dataset.drop(columns=["diagnosis(1=m, 0=b)"])
```

```python
y = dataset["diagnosis(1=m, 0=b)"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
```

```python
#training the data
model.add(tf.keras.layers.Dense(256, input_shape=(x_train.shape[1],), activation='sigmoid'))
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
#Dense is basically default neurons
#Lastly our final value is going to be only one single value that is 0 or 1
```

```python
#testing the data
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#this is to compile our mode
#optimizer used to determine how the weights of the neurons are finr tune with the data
#loss function because we use Binary classification so, metric called binary good for categorical stuff
#We want to correctly classify the results as possible so metric used is accuracy
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```
